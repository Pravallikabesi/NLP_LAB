```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Your sentence
text = "The cat sits on the mat."

# Process the text
doc = nlp(text)

# Print each word with its POS tag
for token in doc:
    print(token.text, "→", token.pos_)
```
