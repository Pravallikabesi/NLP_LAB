```python
from nltk.stem import PorterStemmer

# Initialize stemmer
ps = PorterStemmer()

words = ["running", "runs", "runner", "easily", "fairly", "studies", "studying"]

print("Stemming results:")
for word in words:
    print(word, "->", ps.stem(word))
```
